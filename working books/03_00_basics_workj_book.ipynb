```python
### alternate to using in place instead, reassign players

players = pd.read_csv(
    path.join(DATA_DIR, '2018-season', 'players.csv')) # player data
players.set_index('playerID')
```

```python
# Undo that change

players.reset_index().head()
```

```python
players_cuba = players.loc[players['birthCountry'] == 'Cuba',
                           ['name', 'height', 'weight', 'birthCountry']]
players_cuba.head()
```

```python
#### Sort a frame by a column

### In this case by weight

players_cuba.sort_values('weight', inplace=True)
players_cuba.head()
```
