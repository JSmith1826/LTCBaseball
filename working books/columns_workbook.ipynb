```python
import pandas as pd
from pandas import DataFrame, Series
from os import path

import numpy as np

```

```python
# Load Data

DATA_DIR ='/Users/Justin/Desktop/ltcwbb-files/data'

pp = pd.read_csv(path.join(DATA_DIR, '100-game-sample','pitches.csv'))
print(pp.shape)
pp.head()
```

```python
pp.columns
```

```python
# Inserts a value
pp['mph_max'] = 105.1

pp[['pitcher','batter','i','o','b','s','pitch_type','mph','mph_max']].head()
```

```python
# Create a new column from doing a math operation on other columns

# here the Height of something is calculated by subtracting the top value from the bottom value

pp['sz_height'] = pp['sz_top'] - pp['sz_bot']
pp[['pitcher', 'batter', 'i', 'o', 'b', 's', 'sz_top', 'sz_bot',
    'sz_height']].head()
```

```python
pp['distance_from_middle_of_plate'] = np.abs(pp['x0'])

pp['ln_spin_rate'] = np.log(pp['spin_rate'])

pp.head()
```

```python
# Create a boolean column based on a value of another column
# in this eaxmple we create a Y/N bianary column of is the pitch a fastball
# pitches that have the string FF in the Pitch type column will be marked true


# Bool columns
pp['is_a_FF'] = (pp['pitch_type'] == 'FF')
pp[['pitcher', 'pitch_type', 'is_a_FF']].sample(5)
```

```python
# other variations on the same thing with various conditions

# Is the pitch a fastball or a slider
pp['is_a_FF_or_SL'] = (pp['pitch_type'] == 'FF') | (pp['pitch_type'] == 'SL')

# creates a new column to mark Fast Fastballs, onces that are over 100 mph. 
# to be true the pitch type has to be FF and mph must be greater than 100
pp['fast_fastball'] = (pp['pitch_type'] == 'FF') & (pp['mph'] >= 100)
pp['is_not_FF_or_SL'] = ~((pp['pitch_type'] == 'FF') | (pp['pitch_type'] == 'SL'))

(pp[['b', 's']] == 0).sample(5)
```

```python

# String Columns
pp['pitcher'].str.upper().sample(5)

pp['batter'].str.replace('.', ' ').sample(5)

(pp['pitcher'] + ' throws a ' + pp['pitch_type'] + ' to ' + pp['batter']).sample(5)

pp['batter'].str.replace('.', ' ').str.lower().sample(5)
```
